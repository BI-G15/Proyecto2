```python
unidades = {}
distintos = {}
for i,row in df_mineria.iterrows():
  if row["recurso_natural"] not in unidades:
    unidades[row["recurso_natural"]] = row["unidad_medida"]
  else:
    if unidades[row["recurso_natural"]] != row["unidad_medida"]:
      distintos[row["recurso_natural"]] = row["unidad_medida"]

print(unidades)
print(distintos)
```

## Limpieza de datos

#### Limpieza verificando los municipios

Guardar la informacion de los municipios en un diccionario

```python
municipios = {} 

for i, row in df_depymun.iterrows():
    municipio = row['MUNICIPIO'].lower().replace("í","i").replace("á","a").replace("é","e").replace("ó","o").replace("ú","u")
    departamento = row['DEPARTAMENTO'].lower().replace("í","i").replace("á","a").replace("é","e").replace("ó","o").replace("ú","u")
    municipios[municipio] = departamento
```

Recorrer dataframes y obtener informacion de mineria unificada por el municipio
